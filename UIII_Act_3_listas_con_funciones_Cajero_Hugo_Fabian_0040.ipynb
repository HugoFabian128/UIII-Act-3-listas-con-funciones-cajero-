5. modificarcajero()
6. borrarcajero()

```python
# Lista global para almacenar los cajeros como diccionarios [1, 2]
cajeros = []

def agregar_cajero():
    # Solicita datos y añade un nuevo diccionario a la lista [3, 4]
    id_c = input("ID del cajero: ")
    nombre = input("Nombre del cajero: ")
    cajeros.append({"id": id_c, "nombre": nombre})
    print("Cajero agregado.")

def listar_cajeros():
    # Recorre la lista y muestra todos los elementos [5, 6]
    if not cajeros:
        print("No hay cajeros registrados.")
    else:
        for c in cajeros:
            print(f"ID: {c['id']} | Nombre: {c['nombre']}")

def buscar_cajero():
    # Busca un cajero por ID y devuelve el objeto si lo encuentra [6, 7]
    id_busqueda = input("ID a buscar: ")
    for c in cajeros:
        if c['id'] == id_busqueda:
            return c
    return None

def leer_cajero():
    # Utiliza la función buscar para mostrar los detalles de uno solo [8, 9]
    c = buscar_cajero()
    if c:
        print(f"Detalles - ID: {c['id']}, Nombre: {c['nombre']}")
    else:
        print("Cajero no encontrado.")

def modificar_cajero():
    # Localiza el cajero y actualiza sus valores en el diccionario [3, 7]
    c = buscar_cajero()
    if c:
        c['nombre'] = input("Nuevo nombre: ")
        print("Cajero modificado.")
    else:
        print("No se pudo modificar.")

def borrar_cajero():
    # Busca el cajero y lo elimina de la lista usando su valor [10]
    c = buscar_cajero()
    if c:
        cajeros.remove(c)
        print("Cajero borrado.")
    else:
        print("No se encontró para borrar.")

# Menú principal interactivo usando un ciclo while infinito [11, 12]
while True:
    print("\n--- MENÚ CAJEROS ---")
    print("1. Agregar | 2. Leer | 3. Listar | 4. Buscar | 5. Modificar | 6. Borrar | 7. Salir")
    opcion = input("Seleccione una opción: ")

    if opcion == '1': agregar_cajero()
    elif opcion == '2': leer_cajero()
    elif opcion == '3': listar_cajeros()
    elif opcion == '4':
        if buscar_cajero(): print("El cajero existe.")
        else: print("El cajero no existe.")
    elif opcion == '5': modificar_cajero()
    elif opcion == '6': borrar_cajero()
    elif opcion == '7': break # Finaliza el ciclo y el programa [11]
    else: print("Opción no válida.")
```
